### numpy 的array操作

#### 2.建立一个一维数组 a 初始化为[4,5,6], (1)输出a 的类型（type）(2)输出a的各维度的大小（shape）(3)输出 a的第一个元素（值为4）

```python
a = np.array([4, 5, 6])
print(f"   (1) 类型: {type(a)}")  
print(f"   (2) 维度大小: {a.shape}")
print(f"   (3) 第一个元素: {a[0]}\n")
```

#### 3.建立一个二维数组 b,初始化为 [ [4, 5, 6],[1, 2, 3]] (1)输出各维度的大小（shape）(2)输出 b(0,0)，b(0,1),b(1,1) 这三个元素（对应值分别为4,5,2）

```python
b = np.array([[4, 5, 6], [1, 2, 3]])
print(f"   (1) 维度大小: {b.shape}")
print(f"   (2) 元素 b[0,0]={b[0,0]}, b[0,1]={b[0,1]}, b[1,1]={b[1,1]}\n")
```

#### 4.  (1)建立一个全0矩阵 a, 大小为 3x3; 类型为整型（提示: dtype = int）(2)建立一个全1矩阵b,大小为4x5;  (3)建立一个单位矩阵c ,大小为4x4; (4)生成一个随机数矩阵d,大小为 3x2.

```python
a_zeros = np.zeros((3, 3), dtype=int)
b_ones = np.ones((4, 5))
c_eye = np.eye(4)
d_rand = np.random.random((3, 2))

print(f"(1) 3x3 全0整型:\n{a_zeros}")
print(f"(2) 4x5 全1:\n{b_ones}")
print(f"(3) 4x4 单位矩阵:\n{c_eye}")
print(f"(4) 3x2 随机矩阵:\n{d_rand}\n")
```

#### 24,画图，y=x*x 其中 x = np.arange(0, 100, 0.1) （提示这里用到  matplotlib.pyplot 库）

```python
x_plot1 = np.arange(0, 100, 0.1)
y_plot1 = x_plot1 * x_plot1

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1) # 左边子图
plt.plot(x_plot1, y_plot1)
plt.title(r'$y = x^2$')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
```

#### 25.画图。画正弦函数和余弦函数， x = np.arange(0, 3 * np.pi, 0.1)(提示：这里用到 np.sin() np.cos() 函数和 matplotlib.pyplot 库)

```python
x_plot2 = np.arange(0, 3 * np.pi, 0.1)
y_sin = np.sin(x_plot2)
y_cos = np.cos(x_plot2)

plt.subplot(1, 2, 2) # 右边子图
plt.plot(x_plot2, y_sin, label='Sin')
plt.plot(x_plot2, y_cos, label='Cos', linestyle='--')
plt.title('Sine and Cosine')
plt.xlabel('x')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```
